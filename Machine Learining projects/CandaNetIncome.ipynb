```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("D:\\SEMINAR\\canda.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("PerCapital Income (in USD)")
plt.title("CPI Preciction")
plt.scatter(df.year,df.CPI,color="red",marker="+")
```

```python
ndf=df.drop('CPI',axis='columns')
ndf
```

```python
cpi=df.CPI
cpi
```

```python
model=linear_model.LinearRegression()
model.fit(ndf,cpi)
```

```python
model.predict([[2020]])
```
